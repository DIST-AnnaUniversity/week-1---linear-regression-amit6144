```python
import numpy as np
#equetion is y=mx+c  
X = np.array([0,3,2,1,5,5,4,3,0,4])
y = np.array([2,8,7,3,13,12,9,9,3,10])
m,c = np.polyfit(X,y, 1)
print("Equetion of Best fit is:  y=",m,"(x)",c)
```
